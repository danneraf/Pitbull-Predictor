```python
#https://stackoverflow.com/questions/30539679/python-read-several-json-files-from-a-folder 
# Define a list to store the data
combined_data = []

# Get a list of all JSON files in the directory
json_files = glob.glob('./*.json')

# Iterate over each JSON file
for file in json_files:
    # Exclude the Pitbull lyrics 
    if file == './Lyrics_Pitbull.json':
        continue
    
    with open(file, 'r') as json_file:
        # Read the data from the current JSON file
        data = json.load(json_file)
        
        # Add the data to the combined_data list
        combined_data.append(data)

# Combined data to a new JSON file 
output_file = './combined_other_artists.json'
with open(output_file, 'w') as json_output:
    json.dump(combined_data, json_output)
```
